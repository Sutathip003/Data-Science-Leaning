# Objects and Classes


```python
class Car():
    max_speed = 120
    
    # class_attribute
    def __init__(self, make, model, color, speed = 0):
        self.make = make
        self.model = model
        self.color = color
        self.speed = speed
        
    # class_methods
    def accelerate(self, acceleration):
        if self.speed + acceleration <= Car.max_speed:
            self.speed += acceleration
        else:
            self.speed = Car.max_speed

    def get_speed(self):
        return self.speed

#Create objects (instances)
car1 = Car("Toyota", "Vios", "Red")

car1.accelerate(30)

# Print curant speed
print(f"{car1.make} {car1.model} is currently at {car1.get_speed()} km/h.")
```

```python

# %matplotlib inline # Show all matplotlib plots directly inside the notebook.

%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.patches as patches

class Circle(object):
    
    # attribute
    def __init__(self, radius, color):
        self.radius = radius
        self.color = color
        
    # method
    def add_radius(self, r):
        self.radius = self.radius + r
        return (self.radius)

    # method
    def drawCircle(self):
        ax = plt.gca()
        circle = patches.Circle((0, 0), radius=self.radius, fc=self.color)
        ax.add_patch(circle)
        plt.axis('scaled')
        plt.show() 

RedCircle = Circle(5, "red")
#dir(RedCircle)

BlueCircle = Circle(3, "Blue")
```

## Ractangle class

```python
class Rectangle(object):

    # Constructor
    def __init__(self, width, height, color):
        self.width = width
        self.height = height
        self.color = color

    # Method
    def drawRectangle(self):
        ax = plt.gca()
        Rect = patches.Rectangle((0, 0),self.width, self.height, fc=self.color)
        ax.add_patch(Rect)
        plt.axis('scaled')
        plt.show()


# create object
blueRect = Rectangle(3,2, "blue")
blueRect.drawRectangle()
```

```python
YellowRect = Rectangle(2, 5, "Yellow")
YellowRect.drawRectangle()
```
